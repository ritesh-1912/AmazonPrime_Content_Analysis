### Variables Description

df0 (Credits Dataset):

df1 (Titles Dataset):

production_countries: A list of countries where the title was produced. Stored as a string representation of a list.

seasons: The total number of seasons available for a TV show. This column is null for movies.

imdb_score: The average user rating for the title on IMDb, on a scale of 1 to 10.
imdb_votes: The total number of votes received by the title on IMDb.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# ------------------------------------------------------------------------------
# STEP 1: IMPORT LIBRARIES
# ------------------------------------------------------------------------------
import warnings
import ast
warnings.filterwarnings('ignore')

print("✓ Libraries imported successfully!")

# ------------------------------------------------------------------------------
# STEP 2: LOAD DATASETS
# ------------------------------------------------------------------------------

# Load the datasets (make sure files are uploaded to Colab)
titles_df = pd.read_csv('titles.csv')
credits_df = pd.read_csv('credits.csv')

print(f"\n✓ Titles dataset loaded: {titles_df.shape[0]} rows, {titles_df.shape[1]} columns")
print(f"✓ Credits dataset loaded: {credits_df.shape[0]} rows, {credits_df.shape[1]} columns")

# ------------------------------------------------------------------------------
# STEP 3: INITIAL EXPLORATION - TITLES DATASET
# ------------------------------------------------------------------------------

print("\n" + "="*80)
print("TITLES DATASET - INITIAL EXPLORATION")
print("="*80)

print("\n--- First 5 Rows ---")
display(titles_df.head())

print("\n--- Last 5 Rows ---")
display(titles_df.tail())

print("\n--- Dataset Info ---")
titles_df.info()

print("\n--- Statistical Summary ---")
display(titles_df.describe())

print("\n--- Missing Values ---")
missing_df = pd.DataFrame({
    'Column': titles_df.columns,
    'Missing_Count': titles_df.isnull().sum().values,
    'Missing_Percentage': (titles_df.isnull().sum().values / len(titles_df) * 100).round(2)
})
display(missing_df[missing_df['Missing_Count'] > 0])

# ------------------------------------------------------------------------------
# STEP 4: INITIAL EXPLORATION - CREDITS DATASET
# ------------------------------------------------------------------------------

print("\n" + "="*80)
print("CREDITS DATASET - INITIAL EXPLORATION")
print("="*80)

print("\n--- First 5 Rows ---")
display(credits_df.head())

print("\n--- Last 5 Rows ---")
display(credits_df.tail())

print("\n--- Dataset Info ---")
credits_df.info()

print("\n--- Missing Values ---")
print(credits_df.isnull().sum())

# ------------------------------------------------------------------------------
# STEP 5: DATA CLEANING - TITLES DATASET
# ------------------------------------------------------------------------------

print("\n" + "="*80)
print("DATA CLEANING - TITLES DATASET")
print("="*80)

# Create a copy for cleaning
titles_clean = titles_df.copy()

# --- Handle Missing Values ---
print("\n--- Handling Missing Values ---")

# 1. Age Certification - Fill with 'Not Rated'
titles_clean['age_certification'].fillna('Not Rated', inplace=True)
print("✓ age_certification: Filled with 'Not Rated'")

# 2. Runtime - Fill with median based on type
median_runtime_movie = titles_clean[titles_clean['type'] == 'MOVIE']['runtime'].median()
median_runtime_show = titles_clean[titles_clean['type'] == 'SHOW']['runtime'].median()
titles_clean.loc[(titles_clean['type'] == 'MOVIE') & (titles_clean['runtime'].isnull()), 'runtime'] = median_runtime_movie
titles_clean.loc[(titles_clean['type'] == 'SHOW') & (titles_clean['runtime'].isnull()), 'runtime'] = median_runtime_show
print(f"✓ runtime: Movies={median_runtime_movie}, Shows={median_runtime_show}")

# 3. Seasons - Fill with 0 for movies
titles_clean.loc[titles_clean['type'] == 'MOVIE', 'seasons'] = 0
print("✓ seasons: Movies filled with 0")

# 4. IMDB Score - Fill with mean
mean_imdb_score = titles_clean['imdb_score'].mean()
titles_clean['imdb_score'].fillna(mean_imdb_score, inplace=True)
print(f"✓ imdb_score: Filled with mean ({mean_imdb_score:.2f})")

# 5. IMDB Votes - Fill with median
median_imdb_votes = titles_clean['imdb_votes'].median()
titles_clean['imdb_votes'].fillna(median_imdb_votes, inplace=True)
print(f"✓ imdb_votes: Filled with median ({median_imdb_votes:.0f})")

# 6. TMDB Popularity - Fill with median
median_tmdb_popularity = titles_clean['tmdb_popularity'].median()
titles_clean['tmdb_popularity'].fillna(median_tmdb_popularity, inplace=True)
print(f"✓ tmdb_popularity: Filled with median ({median_tmdb_popularity:.2f})")

# 7. TMDB Score - Fill with median
median_tmdb_score = titles_clean['tmdb_score'].median()
titles_clean['tmdb_score'].fillna(median_tmdb_score, inplace=True)
print(f"✓ tmdb_score: Filled with median ({median_tmdb_score:.2f})")

# --- Parse List-like Columns ---
print("\n--- Parsing List Columns ---")

def safe_literal_eval(val):
    """Safely evaluate string representation of list"""
    try:
        return ast.literal_eval(val) if pd.notna(val) else []
    except:
        return []

# Parse genres
titles_clean['genres_list'] = titles_clean['genres'].apply(safe_literal_eval)
print("✓ Created 'genres_list'")

# Parse production countries
titles_clean['production_countries_list'] = titles_clean['production_countries'].apply(safe_literal_eval)
print("✓ Created 'production_countries_list'")

# --- Feature Engineering ---
print("\n--- Feature Engineering ---")

# 1. Primary genre
titles_clean['primary_genre'] = titles_clean['genres_list'].apply(lambda x: x[0] if len(x) > 0 else 'Unknown')
print("✓ Created 'primary_genre'")

# 2. Number of genres
titles_clean['num_genres'] = titles_clean['genres_list'].apply(len)
print("✓ Created 'num_genres'")

# 3. Primary country
titles_clean['primary_country'] = titles_clean['production_countries_list'].apply(lambda x: x[0] if len(x) > 0 else 'Unknown')
print("✓ Created 'primary_country'")

# 4. Number of countries
titles_clean['num_countries'] = titles_clean['production_countries_list'].apply(len)
print("✓ Created 'num_countries'")

# 5. Decade
titles_clean['decade'] = (titles_clean['release_year'] // 10) * 10
print("✓ Created 'decade'")

# 6. Age group categorization
def categorize_age(cert):
    if pd.isna(cert) or cert == 'Not Rated':
        return 'Not Rated'
    elif cert in ['G', 'TV-Y', 'TV-G']:
        return 'Kids'
    elif cert in ['PG', 'TV-PG', 'TV-Y7']:
        return 'Family'
    elif cert in ['PG-13', 'TV-14']:
        return 'Teens'
    elif cert in ['R', 'TV-MA', 'NC-17']:
        return 'Adults'
    else:
        return 'Other'

titles_clean['age_group'] = titles_clean['age_certification'].apply(categorize_age)
print("✓ Created 'age_group'")

# 7. Rating category
def categorize_rating(score):
    if pd.isna(score):
        return 'Unknown'
    elif score >= 8.0:
        return 'Excellent (8.0+)'
    elif score >= 7.0:
        return 'Good (7.0-7.9)'
    elif score >= 6.0:
        return 'Average (6.0-6.9)'
    else:
        return 'Below Average (<6.0)'

titles_clean['rating_category'] = titles_clean['imdb_score'].apply(categorize_rating)
print("✓ Created 'rating_category'")

# 8. Content era (for trend analysis)
def categorize_era(year):
    if year < 1950:
        return 'Classic (Before 1950)'
    elif year < 1980:
        return 'Golden Age (1950-1979)'
    elif year < 2000:
        return 'Modern Era (1980-1999)'
    elif year < 2010:
        return '2000s'
    elif year < 2020:
        return '2010s'
    else:
        return '2020s'

titles_clean['content_era'] = titles_clean['release_year'].apply(categorize_era)
print("✓ Created 'content_era'")

# 9. Runtime category
def categorize_runtime(runtime, content_type):
    if content_type == 'MOVIE':
        if runtime < 90:
            return 'Short (<90 min)'
        elif runtime < 120:
            return 'Standard (90-120 min)'
        else:
            return 'Long (120+ min)'
    else:  # SHOW
        if runtime < 30:
            return 'Short Episode (<30 min)'
        elif runtime < 60:
            return 'Standard Episode (30-60 min)'
        else:
            return 'Long Episode (60+ min)'

titles_clean['runtime_category'] = titles_clean.apply(lambda x: categorize_runtime(x['runtime'], x['type']), axis=1)
print("✓ Created 'runtime_category'")

# --- Handle Outliers ---
print("\n--- Detecting Outliers ---")

# IQR method for runtime
Q1 = titles_clean['runtime'].quantile(0.25)
Q3 = titles_clean['runtime'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

titles_clean['is_runtime_outlier'] = ((titles_clean['runtime'] < lower_bound) |
                                       (titles_clean['runtime'] > upper_bound))
outlier_count = titles_clean['is_runtime_outlier'].sum()
print(f"✓ Detected {outlier_count} runtime outliers (flagged but kept)")

print(f"\n✓ Titles cleaning complete! Shape: {titles_clean.shape}")

# ------------------------------------------------------------------------------
# STEP 6: DATA CLEANING - CREDITS DATASET
# ------------------------------------------------------------------------------

print("\n" + "="*80)
print("DATA CLEANING - CREDITS DATASET")
print("="*80)

# Create a copy for cleaning
credits_clean = credits_df.copy()

# --- Handle Missing Values ---
print("\n--- Handling Missing Values ---")
credits_clean['character'].fillna('N/A', inplace=True)
print("✓ character: Filled with 'N/A'")

# --- Aggregations for Analysis ---
print("\n--- Creating Aggregations ---")

# Credits per title
credits_per_title = credits_clean.groupby('id').size().reset_index(name='total_credits')
print("✓ Created 'credits_per_title'")

# Actors per title
actors_per_title = credits_clean[credits_clean['role'] == 'ACTOR'].groupby('id').size().reset_index(name='total_actors')
print("✓ Created 'actors_per_title'")

# Directors per title
directors_per_title = credits_clean[credits_clean['role'] == 'DIRECTOR'].groupby('id').size().reset_index(name='total_directors')
print("✓ Created 'directors_per_title'")

# Top actors
top_actors = credits_clean[credits_clean['role'] == 'ACTOR']['name'].value_counts().head(20).reset_index()
top_actors.columns = ['name', 'count']
print("✓ Created 'top_actors'")

# Top directors
top_directors = credits_clean[credits_clean['role'] == 'DIRECTOR']['name'].value_counts().head(20).reset_index()
top_directors.columns = ['name', 'count']
print("✓ Created 'top_directors'")

print(f"\n✓ Credits cleaning complete! Shape: {credits_clean.shape}")

# ------------------------------------------------------------------------------
# STEP 7: MERGE DATASETS
# ------------------------------------------------------------------------------

print("\n" + "="*80)
print("MERGING DATASETS")
print("="*80)

# Merge credit counts with titles
titles_with_credits = titles_clean.merge(credits_per_title, on='id', how='left')
titles_with_credits = titles_with_credits.merge(actors_per_title, on='id', how='left')
titles_with_credits = titles_with_credits.merge(directors_per_title, on='id', how='left')

# Fill NaN with 0 for titles without credits
titles_with_credits['total_credits'].fillna(0, inplace=True)
titles_with_credits['total_actors'].fillna(0, inplace=True)
titles_with_credits['total_directors'].fillna(0, inplace=True)

# Convert to int
titles_with_credits['total_credits'] = titles_with_credits['total_credits'].astype(int)
titles_with_credits['total_actors'] = titles_with_credits['total_actors'].astype(int)
titles_with_credits['total_directors'] = titles_with_credits['total_directors'].astype(int)

print(f"✓ Merged successfully! Final shape: {titles_with_credits.shape}")

# ------------------------------------------------------------------------------
# STEP 8: CREATE ANALYSIS-READY DATAFRAMES
# ------------------------------------------------------------------------------

print("\n" + "="*80)
print("CREATING ANALYSIS-READY DATAFRAMES")
print("="*80)

# 1. Genre analysis dataframe
all_genres = titles_clean.explode('genres_list')
genre_counts = all_genres['genres_list'].value_counts().reset_index()
genre_counts.columns = ['genre', 'count']
print("✓ Created 'genre_counts'")

# 2. Country analysis dataframe
all_countries = titles_clean.explode('production_countries_list')
country_counts = all_countries['production_countries_list'].value_counts().head(20).reset_index()
country_counts.columns = ['country', 'count']
print("✓ Created 'country_counts'")

# 3. Yearly trends dataframe
yearly_trends = titles_clean.groupby(['release_year', 'type']).size().reset_index(name='count')
print("✓ Created 'yearly_trends'")

# 4. Decade trends dataframe
decade_trends = titles_clean.groupby(['decade', 'type']).size().reset_index(name='count')
print("✓ Created 'decade_trends'")

# 5. Rating distribution dataframe
rating_dist = titles_clean.groupby('rating_category').agg({
    'imdb_score': 'mean',
    'imdb_votes': 'mean',
    'title': 'count'
}).reset_index()
rating_dist.columns = ['rating_category', 'avg_score', 'avg_votes', 'count']
print("✓ Created 'rating_dist'")

# 6. Genre-Rating correlation
genre_rating = all_genres.groupby('genres_list').agg({
    'imdb_score': 'mean',
    'imdb_votes': 'mean',
    'title': 'count'
}).reset_index()
genre_rating.columns = ['genre', 'avg_score', 'avg_votes', 'count']
genre_rating = genre_rating[genre_rating['count'] >= 10].sort_values('avg_score', ascending=False)
print("✓ Created 'genre_rating'")

# ------------------------------------------------------------------------------
# STEP 9: FINAL SUMMARY
# ------------------------------------------------------------------------------

print("\n" + "="*80)
print("DATA WRANGLING SUMMARY")
print("="*80)

print("\n--- DATASET SIZES ---")
print(f"Titles (cleaned): {titles_clean.shape}")
print(f"Credits (cleaned): {credits_clean.shape}")
print(f"Merged dataset: {titles_with_credits.shape}")

print("\n--- KEY STATISTICS ---")
print(f"Total Titles: {len(titles_clean)}")
print(f"  - Movies: {len(titles_clean[titles_clean['type'] == 'MOVIE'])} ({len(titles_clean[titles_clean['type'] == 'MOVIE'])/len(titles_clean)*100:.1f}%)")
print(f"  - Shows: {len(titles_clean[titles_clean['type'] == 'SHOW'])} ({len(titles_clean[titles_clean['type'] == 'SHOW'])/len(titles_clean)*100:.1f}%)")
print(f"\nUnique Genres: {titles_clean['genres_list'].explode().nunique()}")
print(f"Unique Countries: {titles_clean['production_countries_list'].explode().nunique()}")
print(f"Year Range: {titles_clean['release_year'].min()} - {titles_clean['release_year'].max()}")
print(f"\nAverage IMDB Score: {titles_clean['imdb_score'].mean():.2f}")
print(f"Average Runtime: {titles_clean['runtime'].mean():.1f} minutes")
print(f"Average Credits per Title: {titles_with_credits['total_credits'].mean():.1f}")

print("\n--- TOP 5 GENRES ---")
print(genre_counts.head())

print("\n--- TOP 5 COUNTRIES ---")
print(country_counts.head())

print("\n--- TOP 5 ACTORS ---")
print(top_actors.head())

print("\n--- TOP 5 DIRECTORS ---")
print(top_directors.head())

print("\n" + "="*80)
print("✓ DATA WRANGLING COMPLETE!")
print("✓ ALL DATAFRAMES ARE READY FOR VISUALIZATION!")
print("="*80)

# ------------------------------------------------------------------------------
# STEP 10: DISPLAY CLEANED DATA SAMPLES
# ------------------------------------------------------------------------------

print("\n--- SAMPLE OF CLEANED TITLES DATA ---")
display(titles_clean[['title', 'type', 'release_year', 'primary_genre', 'primary_country',
                       'imdb_score', 'rating_category', 'age_group']].head(10))

print("\n--- SAMPLE OF MERGED DATA WITH CREDITS ---")
display(titles_with_credits[['title', 'type', 'imdb_score', 'total_credits',
                              'total_actors', 'total_directors']].head(10))

print("\n" + "="*80)
print("READY FOR VISUALIZATION AND ANALYSIS!")
print("="*80)
print("\nAvailable DataFrames:")
print("1. titles_clean - Main cleaned titles dataset")
print("2. credits_clean - Cleaned credits dataset")
print("3. titles_with_credits - Merged dataset with credit counts")
print("4. genre_counts - Genre distribution")
print("5. country_counts - Country distribution")
print("6. yearly_trends - Yearly content trends")
print("7. decade_trends - Decade trends")
print("8. rating_dist - Rating distribution")
print("9. genre_rating - Genre-wise ratings")
print("10. top_actors - Top 20 actors")
print("11. top_directors - Top 20 directors")
````

### What all manipulations have you done and insights you found?

````python
# Chart - 1 visualization code
# ==============================================================================
# CHART 1: CONTENT TYPE DISTRIBUTION (MOVIE vs SHOW)
# ==============================================================================
# This visualization shows the distribution of Movies vs TV Shows on Amazon Prime
# Objective: Understand content diversity - What type of content dominates the platform?

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (16, 6)

# Create figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Color palette
colors = ['#FF6B6B', '#4ECDC4']

# ---------------------------
# Subplot 1: Pie Chart
# ---------------------------

# Get value counts
content_counts = titles_clean['type'].value_counts()

# Create pie chart
wedges, texts, autotexts = axes[0].pie(
    content_counts.values,
    labels=content_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    explode=(0.05, 0.05),  # Slightly separate the slices
    shadow=True,
    textprops={'fontsize': 14, 'weight': 'bold'}
)

# Enhance autotext (percentage labels)
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(16)

# Add title
axes[0].set_title('Content Type Distribution on Amazon Prime\n(Movies vs Shows)',
                   fontsize=16, weight='bold', pad=20)

# Add counts in the center (optional enhancement)
total = content_counts.sum()
axes[0].text(0, 0, f'Total\n{total:,}\nTitles',
             ha='center', va='center', fontsize=12, weight='bold')

# ---------------------------
# Subplot 2: Bar Chart
# ---------------------------

# Create bar chart
bars = axes[1].bar(
    content_counts.index,
    content_counts.values,
    color=colors,
    edgecolor='black',
    linewidth=1.5,
    alpha=0.8
)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    axes[1].text(
        bar.get_x() + bar.get_width() / 2.,
        height,
        f'{int(height):,}\n({height/total*100:.1f}%)',
        ha='center',
        va='bottom',
        fontsize=12,
        weight='bold'
    )

# Customize the bar chart
axes[1].set_xlabel('Content Type', fontsize=14, weight='bold')
axes[1].set_ylabel('Number of Titles', fontsize=14, weight='bold')
axes[1].set_title('Count of Movies vs TV Shows', fontsize=16, weight='bold', pad=20)
axes[1].tick_params(axis='both', labelsize=12)
axes[1].grid(axis='y', alpha=0.3)

# Add y-axis formatting
axes[1].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):,}'))

# Overall title
fig.suptitle('Amazon Prime Video: Content Type Analysis',
             fontsize=18, weight='bold', y=1.02)

# Adjust layout
plt.tight_layout()
plt.show()

# ---------------------------
# Print Key Insights
# ---------------------------

print("\n" + "="*80)
print("CHART 1: KEY INSIGHTS - CONTENT TYPE DISTRIBUTION")
print("="*80)

movie_count = content_counts['MOVIE']
show_count = content_counts['SHOW']
movie_pct = (movie_count / total) * 100
show_pct = (show_count / total) * 100

print(f"\n📊 Total Content on Amazon Prime: {total:,} titles")
print(f"\n🎬 Movies: {movie_count:,} ({movie_pct:.1f}%)")
print(f"📺 TV Shows: {show_count:,} ({show_pct:.1f}%)")
print(f"\n💡 Insight: Movies dominate Amazon Prime's library, making up {movie_pct:.1f}% of all content.")
print(f"   The platform has {movie_count/show_count:.1f}x more movies than TV shows.")

# Additional statistics
print(f"\n📈 Additional Statistics:")
print(f"   - Ratio of Movies to Shows: {movie_count/show_count:.2f}:1")
print(f"   - For every 1 TV Show, there are approximately {int(movie_count/show_count)} Movies")

print("\n🎯 Business Implication:")
if movie_pct > 70:
    print("   - Amazon Prime is heavily movie-focused, which may appeal to audiences")
    print("     looking for feature-length content rather than episodic series.")
else:
    print("   - Amazon Prime has a balanced mix of movies and shows.")

print("="*80)
````
